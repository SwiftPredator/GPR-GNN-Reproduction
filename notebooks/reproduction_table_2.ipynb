```python
import pandas as pd
import math

df = pd.read_csv("../src/results_raw_table_2/table2_nogrid_100.csv", header= None)
```

```python
# taken from t table given num of samples and desired confidence
t_distribuation = 1.984 
num_of_samples = 100
```

```python
df_sparse = df


data = []
for model in df_sparse[0].unique():
    
    row = [model]
    for dataset in df_sparse[1].unique():
        mean_acc = df_sparse[(df_sparse[0]==model) & (df_sparse[1]==dataset)][7]
        std = df_sparse[(df_sparse[0]==model) & (df_sparse[1]==dataset)][8]
        std_error = std / math.sqrt(num_of_samples)
        ci = std_error * t_distribuation

        row.append(mean_acc.values[0].round(2).astype(str) + " + " + ci.values[0].round(2).astype(str))
    data.append(row)
    row = []

table_sparse = pd.DataFrame(data, columns= ["model", 'cora', 'citeseer', 'pubmed', 'computers', 'photo', 'chameleon', 'film', 'squirrel',
 'texas', 'cornell'])
```

```python
print(table_sparse.to_latex(index=False))
table_sparse
```

```python
phi_dfs = []
phis = [-1.0, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1]
for phi in phis:
    phi_csv = pd.read_csv(f"../src/results_raw_cSBM/cSBM_data_phi_{phi}.csv", header=None)
    phi_dfs.append(phi_csv)

combined = pd.concat(phi_dfs, axis=0)
```

```python
def generate_table(df_input):
    data = []
    for model in df_input[0].unique():
        
        row = [model]
        for dataset in df_input[1].unique():
            mean_acc = df_input[(df_input[0]==model) & (df_input[1]==dataset)][7]
            std = df_input[(df_input[0]==model) & (df_input[1]==dataset)][8]
            std_error = std / math.sqrt(num_of_samples)
            ci = std_error * t_distribuation

            row.append(mean_acc.values[0].round(2).astype(str) + " + " + ci.values[0].round(2).astype(str))

        data.append(row)
        row = []

    return pd.DataFrame(data, columns= ["model", '-1', '-0.75', '-0.5', '-0.25', '0', '0.25', '0.5', '0.75',
    '1'])
```

```python
df_dense = combined[combined[2]==0.6]
table_dense = generate_table(df_dense)

print(table_dense.to_latex(index=False))
table_dense.to_csv("cSBM_dense.csv")
```

```python
df_sparse = combined[combined[2]==0.025]
table_dense = generate_table(df_dense)

print(table_dense.to_latex(index=False))
table_dense.to_csv("../data/cSBM_sparse.csv")
```
