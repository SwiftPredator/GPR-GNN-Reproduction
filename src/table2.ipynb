```python
t_distribuation = 1.984 
num_of_samples = 100

df_sparse = df
print(df_sparse[1].unique())
data = []
for model in df_sparse[0].unique():
    
    row = [model]
    for dataset in df_sparse[1].unique():
        mean_acc = df_sparse[(df_sparse[0]==model) & (df_sparse[1]==dataset)][7]
        std = df_sparse[(df_sparse[0]==model) & (df_sparse[1]==dataset)][8]
        std_error = std / math.sqrt(num_of_samples)
        ci = std_error * t_distribuation

        row.append(mean_acc.values[0].round(2).astype(str) + " + " + ci.values[0].round(2).astype(str))
    data.append(row)
    row = []

table_sparse = pd.DataFrame(data, columns= ["model", 'cora', 'citeseer', 'pubmed', 'computers', 'photo', 'chameleon', 'film', 'texas',
 'cornell', 'squirrel'])
```
